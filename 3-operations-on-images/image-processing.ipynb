```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('../images/car-image.jpeg')
px = img[100, 100]
print (px)

# access only blue pixel
blue = img[100, 100, 0]
print (blue )
```

```python
# modify one pixel to be white at 100,100
img[100, 100] = [255,255,255]

# cv2 uses BGR convert it to RGB to get normal colours
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

You can see that the pixel at 100,100 is now white

```python
# It returns a tuple of number of rows, columns and channels (if image is color):
shape = img.shape
print ('Shape:', shape)

```
